# 2) Check shape & missing values 

```python
ds1 = df1.copy()
ds2 = df2.copy()
```

```python
# checking shape and missing values
print("Dataset 1 shape:", ds1.shape)
print("\nMissing values in Dataset 1:\n")
print(ds1.isnull().sum())
```

```python
print("Dataset 2 shape:", ds2.shape)
print("\nMissing values in Dataset 2:\n")
print(ds2.isnull().sum())
```

# 3) Handle missing values in TARGET

```python
# Count nulls in Target before dropping
print("Null values in Target before:", ds2['Target'].isnull().sum())
```

```python
# Drop rows where Target is null
ds2 = ds2.dropna(subset=['Target'])
```

```python
print("Null values in Target after:", ds2['Target'].isnull().sum())
print("Dataset 2 shape after dropping Target nulls:", ds2.shape)
```

# 4) Identify Numerical & Categorical Columns

```python
cat_cols_ds1 = ds1.select_dtypes(include='object').columns
num_cols_ds1 = ds1.select_dtypes(exclude='object').columns
cat_cols_ds2 = ds2.select_dtypes(include='object').columns
num_cols_ds2 = ds2.select_dtypes(exclude='object').columns
```

```python
print("Category columns in dataset1")
print(len(cat_cols_ds1))
for col in cat_cols_ds1:
    print(col)
print("=============================")
print("Category columns in dataset2")
print(len(cat_cols_ds2))
for col in cat_cols_ds2:
    print(col)
print("=============================")
print("Numerical columns in dataset1")
print(len(num_cols_ds1))
for col in num_cols_ds1:
    print(col)
print("=============================")
print("Numerical columns in dataset2")
print(len(num_cols_ds2))
for col in num_cols_ds2:
    print(col)
```

# 5) Handle Missing Values (Numerical) -> Mean Imputation

```python
for col in num_cols_ds1:
    mean_value = ds1[col].mean()
    ds1[col] = ds1[col].fillna(mean_value)
```

```python
for col in num_cols_ds2:
    mean_value = ds2[col].mean()
    ds2[col] = ds2[col].fillna(mean_value)
```

# 6) Handle Missing Values (Categorical) -> Mode Imputation

```python
for col in cat_cols_ds1:
    mode_value = ds1[col].mode()[0]
    ds1[col] = ds1[col].fillna(mode_value)
```

```python
ds2 = ds2.dropna(subset=['Target'])
```

```python
for col in cat_cols_ds2:
    mode_value = ds2[col].mode()[0]
    ds2[col] = ds2[col].fillna(mode_value)
```

# 7) Verifying missing values after cleaning

```python
print("Missing values in Dataset 1 after cleaning:\n")
print(ds1.isnull().sum())

print("\nMissing values in Dataset 2 after cleaning:\n")
print(ds2.isnull().sum())
```

# 8) Merging both cleaned datasets

```python
final_df = pd.concat([ds1, ds2], axis=1)

print("Final dataset shape:", final_df.shape)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select a couple of numerical columns for illustration
cols_to_plot = ['Age at enrollment', 'Curricular units 1st sem (grade)']

plt.figure(figsize=(10,4))

for i, col in enumerate(cols_to_plot):
    plt.subplot(1, len(cols_to_plot), i+1)
    sns.histplot(final_df[col], kde=True, color='skyblue')
    plt.title(f"{col} (Before Scaling)")

plt.tight_layout()
plt.show()
```

# 9) Apply Min Max Scaling

```python
num_cols = final_df.select_dtypes(include=[np.number]).columns
```

```python
num_cols
```

```python
# Define continuous numeric columns for scaling only
continuous_cols = [
    'Age at enrollment',
    'Curricular units 1st sem (credited)',
    'Curricular units 1st sem (enrolled)',
    'Curricular units 1st sem (evaluations)',
    'Curricular units 1st sem (approved)',
    'Curricular units 1st sem (grade)',
    'Curricular units 1st sem (without evaluations)',
    'Curricular units 2nd sem (credited)',
    'Curricular units 2nd sem (enrolled)',
    'Curricular units 2nd sem (evaluations)',
    'Curricular units 2nd sem (approved)',
    'Curricular units 2nd sem (grade)',
    'Curricular units 2nd sem (without evaluations)',
    'Unemployment rate',
    'Inflation rate',
    'GDP'
]
```

```python
for col in continuous_cols:
    min_val = final_df[col].min()
    max_val = final_df[col].max()
    final_df[col] = (final_df[col] - min_val) / (max_val - min_val)
```

```python
print(final_df[continuous_cols].describe().T[['min','max']])
```

```python
cols_to_plot = ['Age at enrollment', 'Curricular units 1st sem (grade)']

plt.figure(figsize=(10,4))

for i, col in enumerate(cols_to_plot):
    plt.subplot(1, len(cols_to_plot), i+1)
    sns.histplot(final_df[col], kde=True, color='salmon')
    plt.title(f"{col} (After Scaling)")

plt.tight_layout()
plt.show()
```
