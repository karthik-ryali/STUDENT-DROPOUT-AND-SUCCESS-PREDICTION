## 1. Dataset Overview

This dataset contains information about **4,424 students** and **35 features** related to:
- Demographic background
- Academic performance (1st and 2nd semester)
- Socio-economic conditions
- Final academic outcome (Target)

The **Target** variable represents the student status:
- **Dropout** – student discontinued studies  
- **Graduate** – student successfully completed the course  
- **Enrolled** – student is still studying  

The dataset is already **clean**, with:
- No missing values
- No duplicate records

So we directly proceed to **exploratory data analysis (EDA)**.
## 2. Descriptive Statistics Summary

From the descriptive statistics, we observe:
- Most features are **numerical and encoded**
- Academic performance variables (grades, approvals, enrollments) show high variation
- Economic indicators such as **Unemployment rate, Inflation rate, and GDP** are included
- Student age at enrollment varies, indicating a mix of traditional and non-traditional students

These statistics help us understand:
- Overall data distribution
- Possible factors influencing student success or dropout

## 3. Data Quality Check

The missing value analysis confirms that:
- **No columns contain missing values**
- The dataset is complete and consistent

Because of this, **no data cleaning or imputation** is required.
We can safely move forward with visualization and analysis.

```python
# Plot distribution of student status (target variable)
plt.figure(figsize=(10, 6))
ax = sns.countplot(
    data=df,
    x="Target",
    order=["Dropout", "Graduate", "Enrolled"]
)

# Add percentage labels above each bar
total = len(df)

for p in ax.patches:
    height = p.get_height()
    ax.text(
        p.get_x() + p.get_width() / 2,
        height + 5,
        f"{height / total * 100:.1f}%",
        ha="center"
    )

# Customize the plot
plt.title("Distribution of Student Status")
plt.xlabel("Student Status")
plt.ylabel("Number of Students")
plt.show()
```

## 4. Distribution of Student Outcomes (Target Variable)

In this section, we analyze the distribution of student outcomes:
- Dropout
- Graduate
- Enrolled

Understanding the target distribution is important to:
- Identify class imbalance
- Understand the overall academic success trend

### Observations:
- The majority of students either **graduate or drop out**
- A smaller proportion remains **currently enrolled**
- This imbalance suggests that dropout prediction is an important problem

This insight motivates further analysis of factors influencing student outcomes.

```python
# Plot age distribution at enrollment
plt.figure(figsize=(10, 6))
sns.histplot(
    df["Age_at_enrollment"],
    bins=20,
    kde=True
)
plt.title("Age_Distribution_at_Enrollment")
plt.xlabel("Age")
plt.ylabel("Number_of_Students")
plt.show()
```

## 7. Academic Performance Analysis – Second Semester

Second semester grades further strengthen the academic trend:
- Students who graduate maintain higher performance
- Dropout students show consistently lower grades

This highlights the importance of **continuous academic monitoring**.

```python
# Analyze debtor status vs student outcome
plt.figure(figsize=(8, 5))
sns.countplot(
    data=df,
    x="Debtor",
    hue="Target"
)

plt.title("Debtor Status vs Student Status")
plt.xlabel("Debtor Status (0 = No, 1 = Yes)")
plt.ylabel("Number of Students")

# Add count labels on bars
for p in plt.gca().patches:
    plt.gca().annotate(
        int(p.get_height()),
        (p.get_x() + p.get_width() / 2, p.get_height()),
        ha="center",
        va="center",
        fontsize=9,
        xytext=(0, 5),
        textcoords="offset points"
    )

# Show legend and plot
plt.legend(title="Student Status")
plt.tight_layout()
plt.show()
```

```python
# Create approval rate for the first semester
df["Approval_Rate_1st_Sem"] = (
    df["Curricular_units_1st_sem_approved"] /
    df["Curricular_units_1st_sem_enrolled"]
)
```

```python
# Compare first semester approval rate by student status
plt.figure(figsize=(10, 6))
sns.boxplot(
    data=df,
    x="Target",
    y="Approval_Rate_1st_Sem"
)

plt.title("First Semester Approval Rate vs Student Status")
plt.xlabel("Student Status")
plt.ylabel("Approval Rate")
plt.show()
```

```python
# Create approval rate for the secong semester
df["Approval_Rate_2nd_Sem"] = (
    df["Curricular_units_2nd_sem_approved"] /
    df["Curricular_units_2nd_sem_enrolled"]
)
```

```python
# Compare second semester approval rate by student status
plt.figure(figsize=(10, 6))
sns.boxplot(
    data=df,
    x="Target",
    y="Approval_Rate_2nd_Sem"
)

plt.title("Second Semester Approval Rate vs Student Status")
plt.xlabel("Student Status")
plt.ylabel("Approval Rate")
plt.show()
```

## 10. Approval Rate Analysis

Approval rate represents the proportion of enrolled courses successfully passed.
It is a strong indicator of:
- Academic consistency
- Risk of dropout

### Key Insights:
- Graduated students show significantly higher approval rates
- Dropout students consistently have lower approval rates
- Approval rate is one of the **strongest predictors of student success**
